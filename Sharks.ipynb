```python
def date(actual):
    actual=actual.replace('Reported','')
    actual=actual.replace('Between','')
    actual=actual.replace(')','')
    actual=actual.replace('(same day as','-')
    actual=actual.replace('"Bitten last weekend','')
    actual=actual.replace('Some time between','')
    actual=actual.replace('Before','')
    
    try:
        actual2=pd.to_datetime(actual)
        return actual2
    except:
        return actual
```

```python
sharks.Date=sharks.Date.apply(date)
```
